```python
Temp=[]
for i in range(0,len(vx_1)):
    dummylist=[]
    for j in range(0,len(vx_1[i])):
        prod=(vx_1[i][j])**2+(vy_1[i][j])**2
        dummylist.append(prod)
    Temp.append((1/32)*np.sum(dummylist))
```

```python
TenK_avg_temp=[]
for i in range(10000,len(Temp)):
    TenK_avg_temp.append(np.average(Temp[10000:i]))
```

```python
#Way more efficient made this too late to use much but would have saved significant time as scales with N
def Time_Average(series):
    avg_series=[]
    avg=series[0]
    avg_series.append(avg)
    for i in range(1,len(series)):
        avg=(1/i)*((i-1)*avg+(series[i]))
        avg_series.append(avg)

    return avg_series
```

```python
test=Time_Average(Temp[10000:])
print(test[-1])
print(TenK_avg_temp[-1])
print(np.std(Temp[10000:]))
```

```python
#plt.plot(test)
plt.plot(test,label=r'$\overline{T}$')
plt.ylim(0.95,1.03)
plt.grid(True)
plt.xlabel('Time Steps Post Equilibrium')
plt.ylabel(r"$\overline{T}$ (Reduced Units)")
plt.title("Time Evolution of Time Averaged Temperature")
plt.yticks(np.arange(0.95,1.03,0.01))
#plt.axhline(y=0.984375,color='black',ls='--',label='Expected from Equipartition with K(t)=')
plt.plot([(1/16)*el for el in Time_Average(k_1[10000:])],ls=':',label='Expected from Equipartition')
#plt.plot([(1/16)*el for el in k_1[10000:]])
plt.legend()
```

```python
#Velocity distribution stuff
vmag=[]
for i in range(len(vx_1)):
    time_step=[]
    for j in range(len(vx_1[i])):
        time_step.append((np.sqrt(vx_1[i][j]**2+vy_1[i][j]**2)))
    vmag.append(time_step)
```

```python
def P(x,A,T):
    return A*x*np.exp((-x**2)/T)
```

```python
v_1=np.asarray(vmag[0:500]).flatten()
v_2=np.asarray(vmag[500:1000]).flatten()
v_3=np.asarray(vmag[1000:1500]).flatten()
v_4=np.asarray(vmag[1500:2000]).flatten()
v_5=np.asarray(vmag[20000:]).flatten()
```
